```python
# feature scaling 
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()

x = sc.fit_transform(x)

x
```

# Model Building 

```python
from sklearn.cluster import DBSCAN
```

```python
# lets go with the default parameter 

db = DBSCAN()

db.fit_predict(x)
```

```python
# evolution metrics of unsupervized learning 
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(x, db.fit_predict(x))
```

```python
# chage the parameter with eps = 1 and min_samples=7  

db = DBSCAN(eps=1,min_samples=7)

db.fit_predict(x)

silhouette_score(x, db.fit_predict(x))
```

```python
# chage the parameter with eps=2,min_samples=3  


db = DBSCAN(eps=2,min_samples=3)

y = db.fit_predict(x)

silhouette_score(x, db.fit_predict(x))
```

```python
# our model performing the 74 % accuracy 
```

```python
df['cluster'] = y
```
