# Model Building 

```python
from sklearn.cluster import DBSCAN
```

```python
db = DBSCAN()

y = db.fit_predict(x)

y
```

```python
plt.scatter(x[:,1],x[:,-1],c= y)
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(x,y)
```

```python
#Analyze the clusters
```

```python
df['Cluster'] = y
df
```

```python
df[df['Cluster'] == 1]
```

```python
df.groupby('Cluster').agg('mean')
```

# Kmeans 

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans()

yp = km.fit_predict(x)

yp
```

```python
#cluster centre
km.cluster_centers_
```

```python
plt.scatter(x[:,0],x[:,1], c = yp, cmap='rainbow')

plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1], c = 'yellow', s = 100)
```

```python
wcss = []

for i in range(1,15):
    km = KMeans(n_clusters=i)
    km.fit_predict(x)
    wcss.append(km.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,15),wcss)
plt.axvline(x = 3, linestyle = '--', color = 'red')
plt.xlabel('K-Values')
plt.ylabel('WCSS')
plt.title('Elbow Graph')
```

```python
km = KMeans(n_clusters=3)
ypred = km.fit_predict(x)
ypred
```

```python
plt.scatter(x[:,0],x[:,1], c = ypred, cmap='rainbow')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1], c = 'yellow', s = 150)
```

```python
silhouette_score(x,ypred)
```

```python
#Analyzing clusters
df['cluster'] = ypred
df
```

```python
df.groupby('cluster').agg('mean')
```
