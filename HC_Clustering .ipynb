```python
df.drop(columns='Univ',inplace=True)
df
```

```python
sns.pairplot(df)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()

x = sc.fit_transform(df)

x
```

```python
from scipy.cluster import hierarchy
```

```python
# Dendrogram base on single linkage 

lk = hierarchy.linkage(x,method='single')
dendrogram = hierarchy.dendrogram(lk)
```

```python
# Dendrogram base on complete linkage 
lk = hierarchy.linkage(x,method='complete')
dendrogram = hierarchy.dendrogram(lk)
```

```python
# Dendrogram base on average linkage 

lk = hierarchy.linkage(x,method='average')
dendrogram = hierarchy.dendrogram(lk)
```

```python
# Dendrogram base on centroid linkage 
lk = hierarchy.linkage(x,method='centroid')
dendrogram = hierarchy.dendrogram(lk)
```

#### Base on these all method we found that the complete linkage is very good method and also by the calculation it is effective that is why most of the tiome we are using the complete linkage 

# Model Building 

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hc = AgglomerativeClustering(n_clusters=2)

ypred = hc.fit_predict(x)
```

```python
df['cluster'] = ypred

df
```

```python
# Analyzing the Clusters 
```

```python
df.groupby('cluster').agg('mean')
```

 we found that the univ whos expenses is very high belong to the 0 cluster and the low expenses uviv that belong to the 1 cluster 

```python
sns.scatterplot(df['Top10'],df['Expenses'],hue=df['cluster'])
```

```python
sns.scatterplot(df['Accept'],df['SFRatio'],hue=df['cluster'])
```
